```python
# import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('hotel_booking.csv', encoding= 'unicode_escape')
```

# Exploratory Data analysis and Data Cleaning

```python
df.head()
```

```python
df.shape
```

```python
df.drop(["name", "email","phone-number","credit_card"], axis=1, inplace= True)
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
```

```python
df.describe(include='object')
```

```python
df.isnull().sum()
```

```python
df.drop(["company","agent"], axis=1, inplace= True)
```

```python
df=df[df["adr"]<5000]
```

# Data Analysis and Visualizations

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
```

```python
cancelled_perc
```

```python
plt.figure(figsize=(5,4))
plt.title('reservation status count')
plt.bar(['not cancelled', 'cancelled'],df['is_canceled'].value_counts(),color='purple',width=0.7)
plt.savefig('reservation status count.jpg')
plt.show()
```

```python
plt.figure(figsize=(7,6))
ax1 = sns.countplot(x='hotel',hue = 'is_canceled',data=df,palette="Blues")
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotel')
plt.xlabel('hotel')
plt.ylabel('reservations number')
plt.legend(['not cancelled','cancelled'])
plt.savefig('Reservation status in different hotel.jpg')
plt.show()
```

```python
resort_hotel=df[df['hotel']=="Resort Hotel"]

resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel1=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel1=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
resort_hotel1
```

```python
city_hotel1
```

```python
plt.figure(figsize=(20,8))
plt.title('average daily rate in city and resort hotel',size=20)
plt.plot(resort_hotel1.index,resort_hotel1["adr"],label="resort_hotel")
plt.plot(city_hotel1.index,city_hotel1["adr"],label="city_hotel")
plt.legend()
plt.savefig('average daily rate in city and resort hotel.jpg')
plt.show()
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(6,5))
ax1=sns.countplot(x='month',hue='is_canceled',data=df,palette="bright")
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('reservation status per month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.savefig('reservation status per month.jpg')
plt.show()
```

```python
df1=df[df['is_canceled']==1].groupby('arrival_date_month')[['adr']].sum().reset_index()

```

```python
df1
```

```python
plt.figure(figsize=(15,10))
plt.title('adr per month',fontsize=20)
plt.bar('arrival_date_month','adr',data=df1)
plt.savefig("adr per month.jpg")
plt.show()
```

```python
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.savefig('top 10 countries with reservation canceled.jpg')
plt.show()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
cancelled_data_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_data_adr.reset_index(inplace=True)
cancelled_data_adr.sort_values('reservation_status_date',inplace=True)
not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_data_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_data_adr.reset_index(inplace=True)
not_cancelled_data_adr.sort_values('reservation_status_date',inplace=True)
plt.figure(figsize=(20,8))
plt.title("average daily rate",size=20)
plt.plot(cancelled_data_adr["reservation_status_date"],cancelled_data_adr['adr'],label='cancelled')
plt.plot(not_cancelled_data_adr["reservation_status_date"],not_cancelled_data_adr['adr'],label='cancelled')

plt.savefig('average daily rate.jpg')
plt.show()
```
